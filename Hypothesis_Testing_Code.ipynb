```python
import scipy.stats as stats

# Accuracies obtained from the test dataset for optical and radar models
radar_accuracies = [DTFacc1, Logacc1, Rfacc1, knnacc1]
optical_accuracies = [DTFacc2, Logacc2, Rfacc2, knnacc2]


# Perform a paired t-test
t_statistic, p_value = stats.ttest_rel(optical_accuracies, radar_accuracies)

# Print the results
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Make a decision based on the p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between optical and radar accuracies.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between optical and radar accuracies.")
```
